#### Q2. How many Movies & TV shows are in the dataset ? showing with bar graph

```python
netflix.head(2)
```

```python
# To group all unique times of a column and show their count
netflix.groupby('type').type.count()
```

```python
# To show the count of all unique values of any column in the face of bar graph
netflix.groupby('type').type.count().plot(kind='bar')
```
